```python
import pandas as pd

import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=[10,5]

import requests

url = "https://ntreis.stats.10kresearch.com/infoserv/s-v1/ibHk-irG"
response = requests.get(url)
```

```python
with open("data.csv", "wb") as f:
    f.write(response.content) 
```

```python
df = pd.read_csv('data.csv',skiprows=9)
```

```python
df.head()
```

```python
plt.plot(df['Date'],df['Entire MLS'])
plt.xticks(rotation=50)
```

```python
df['MOM_change']=df['Entire MLS'].pct_change()
```
